```python
import matplotlib.pyplot as plt

# Data
values = [5, 8, 9, 4, 1, 6, 7, 2, 3, 8]

# Create axes
ax = plt.axes()

# Set axis limits
ax.set_xlim([0, 50])
ax.set_ylim([-10, 10])

# Customize ticks
ax.set_xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
ax.set_yticks([-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])

# Enable grid
ax.grid()

# Plot the data
plt.plot(range(1, 11), values)
plt.show()
```

```python
values1 = [2,4,3,1,5,6]
values2 = [3,5,1,5,7,9]
plt.plot(range(6),values1,c='g',ls='--',lw=1,marker='>',label='hello')
plt.plot(range(6),values2,c='r',ls='-',lw=2,marker='o',label='there')
plt.xlabel('roll number')
plt.ylabel('cpi')
plt.annotate(
    "Important Point",        # The text
    xy=(4, 7),                 # Point to annotate
    xytext=(4, 6.5),             # Position of the text
    arrowprops=dict(arrowstyle="->", color='red')  # Add an arrow
)
plt.legend(loc=4,title='title',fontsize=10)
plt.show()
```

# Pie chart

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']
explode = [0, 0.1, 0, 0]  # Highlight Python

# Create Pie Chart
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

plt.title("Programming Language Popularity")
plt.legend(title="Languages", loc="upper left")
plt.show()
```
